```python
# Preprocessing the data for multiple linear regression 
# Calculate the profit column
# Creating df_inf variable exclusively for the purpose of multiple linear regression
df_inf = df_cleaned1.copy()
df_inf['profit']  = df_inf['value_sales'] - df_inf['value_cost']
continuous_vars = ['value_sales', 'value_cost', 'value_quantity', 'profit']


# Remove NOR - Normal Order, NOH - Normal Order Head Office Sales if <0 
df_inf = df_inf[~((df_inf['order_type_code'].isin(['NOR','NOH'])) & ((df_inf['value_sales'] < 0) | (df_inf['value_cost'] < 0) | (df_inf['value_quantity'] < 0)))]

# order type code for Credit - credit transaction will result in negative value_sales, value_cost, value_quantity due to the credit nature
# Whichever transaction that has positive result will be error and will be removed
credit_order_type = ['CDG','CRR','CRD','CPR','CRP','ZCG','ZCR']

# First remove the credit type that has positive values
df_inf = df_inf[~((df_inf['order_type_code'].isin(credit_order_type)) & ((df_inf['value_sales'] > 0) | (df_inf['value_cost'] > 0) | (df_inf['value_quantity'] > 0)))]

# Then remove the credit type that has negative values for profit and double check Value Sales, Value Cost, and Value Quantity 
df_inf = df_inf[
    (df_inf['value_quantity'] > 0) & 
    (df_inf['value_sales'] != 0) & 
    (df_inf['value_cost'] != 0) &
    (df_inf['profit'] > 0) 
]

# Calculate the value of skewness and visualize data distribution with histogram
skewness_values = df_inf[continuous_vars].skew()
print("Skewness before transformation:")
print(skewness_values)

# Data is highly skewed, we do the logarithm transformation to reduce the skewness, log1p is used to handle zero values if  any
df_inf['log_value_sales'] = np.log1p(df_inf['value_sales'])
df_inf['log_value_cost'] = np.log1p(df_inf['value_cost'])
df_inf['log_value_quantity'] = np.log1p(df_inf['value_quantity'])
df_inf['log_profit'] = np.log1p(df_inf['profit'])

# Check the skewness after transformation
skewness_after_transformation = df_inf[['log_value_sales', 'log_value_cost', 'log_value_quantity','log_profit']].skew()
print("Skewness after transformation:")
print(skewness_after_transformation)
```

### Insight 3: Impact of Currency on Profit Margins

```python
# Q: Which currency markets deliver the highest true profit margin per dollar sold?
# Method:
#   1) Remove zero-sales orders to avoid div/0
#   2) Sum profit & sales by currency
#   3) Calculate weighted_margin = total_profit/total_sales Ã 100
#   4) Print the results and plot as a sorted horizontal bar chart



# Filter and compute profit
df_cleaned3['profit'] = df_cleaned3['value_sales'] - df_cleaned3['value_cost']
df = df_cleaned3[df_cleaned3['value_sales'] > 0].copy()



# Aggregate totals by currency
summary = df.groupby('currency').agg(total_profit=('profit','sum'),total_sales =('value_sales','sum'))

# Compute weighted profit margin (%) and round
summary['weighted_margin'] = (summary['total_profit'] / summary['total_sales'] * 100).round(1)

# Display the table of weighted margins
print("Weighted Profit Margin by Currency (%):")
print(summary['weighted_margin'].sort_values(), "\n")

# Plot sorted horizontal bar chart
m = summary['weighted_margin'].sort_values()
colors = ['pink','lightsalmon','lightcoral','darkred']  

fig, ax = plt.subplots(figsize=(6, 4))
m.plot(kind='barh',ax=ax,color=colors[:len(m)])
ax.set_title('Weighted Profit Margin by Currency', weight='bold')
ax.set_xlabel('Profit Margin (%)')
ax.set_ylabel('Currency')
ax.grid(axis='x', linestyle='--', alpha=0.5)

# Annotate exact percentages on bars
for i, v in enumerate(m):ax.text(v + 1, i, f"{v:.1f}%", va='center')

plt.tight_layout()
plt.show()
```

**Result of Analysis:** <br>
Our weightedâmargin chart shows that sales billed in EUR deliver by far the highest true profit per dollar soldâaround 95.7 %. Next comes NZD at 70.0 %, then AUS at 64.7 %. By comparison, USDâdenominated orders yield the lowest margin (only 34.8 %). In other words, every dollar of revenue in Europe is almost entirely profit, whereas U.S. sales eat up nearly twoâthirds of their revenue in costs.

**Value for Management:** <br>
With these numbers in hand, leadership can prioritize the currencies and geographies that maximize our bottom line. For example, since European sales pack the biggest profit punch, we might invest more in marketing, localized offerings, or expanding our product range in EUR markets. Conversely, the low margin in USD suggests we need to revisit pricing, negotiate better supplier contracts, or tighten cost controls in the U.S. to bring that figure up. Overall, by doubling down on high-margin regions (EUR, NZD, AUS) and fixing inefficiencies where margins lag (USD), we can boost company profitability without necessarily raising prices across the board.

# Answer to Section 3: Test Sub Sample Differences (all the codes in one cell - do not insert extra cells) 

## section 3.1 (all the codes in one cell - do not insert extra cells) 

```python
## Question 1: Do âHighâperformingâ ABC items (A + B) differ in mean profit' from âLowâperformingâ items (C, D, E, G)?
# Grouping âHighâ performers (classes A & B) vs âLowâ performers (classes C, D, E, G) to see if thereâs a real margin advantage to stocking your topâtier SKUs.

# H0: There is no significant difference in average profit between High (A+B) and Low (C, D, E, G) ABC classes.
# H1: There is a significant difference in average profit between High (A+B) and Low (C, D, E, G) ABC classes.

from scipy import stats

# compute profit 
df_cleaned3['profit'] = df_cleaned3['value_sales'] - df_cleaned3['value_cost']


# Filter out rows with zero or negative profit
df_cleaned3 = df_cleaned3[df_cleaned3['profit'] > 0]


# Separate the data into High and Low ABC groups
high_profit = df_cleaned3[df_cleaned3['abc_class_code'].isin(['A','B'])]['profit']
low_profit  = df_cleaned3[df_cleaned3['abc_class_code'].isin(['C','D','E','G'])]['profit']

# Perform Welchâs t-test
t_stat, p_value = stats.ttest_ind(high_profit, low_profit, equal_var=False)

# Print the results
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")

# Conclusion at Î± = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Significant difference in mean profit between High and Low ABC classes.")
else:
    print("Fail to reject H0: No significant difference in mean profit between these groups.")

# Compute and print the group means
mean_high = high_profit.mean()
mean_low  = low_profit.mean()
print(f"Mean profit: (High ABC A+B): {mean_high}")
print(f"Mean profit: (Low  ABC C,D,E,G): {mean_low}")

```

```python
## Question 2: Is the average units per invoice (value_quantity) different between Australian and New Zealand districts?
# Comparing Domestic (Australia) vs. NZ order sizes guides inventory planning and regional promotions.

# H0: There is no significant difference in mean value_quantity between Australian and New Zealand districts.
# H1: There is a significant difference in mean value_quantity between Australian and New Zealand districts.

from scipy import stats

# Define the district groups
aus_codes = [200, 300, 400, 600, 500, 410, 310, 210, 510]  # major Australian districts
nz_codes  = [530, 535, 540, 520, 545]                     # New Zealand districts

aus_qty = df_cleaned3[df_cleaned3['customer_district_code'].isin(aus_codes)]['value_quantity']
nz_qty  = df_cleaned3[df_cleaned3['customer_district_code'].isin(nz_codes)]['value_quantity']

# Perform Welchâs t-test
t_stat, p_value = stats.ttest_ind(aus_qty, nz_qty, equal_var=False)

# Output the results
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")

# Conclusion at Î± = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Significant difference in average units per invoice between Australia and NZ.")
else:
    print("Fail to reject H0: No significant difference in average units per invoice between these regions.")

# Compute and print the group means
mean_aus = aus_qty.mean()
mean_nz  = nz_qty.mean()
print(f"Mean units per invoice (Australia): {mean_aus:.2f}")
print(f"Mean units per invoice (New Zealand): {mean_nz:.2f}")
```

```python
#### Question 1:  
#How does the item class category impact profitability, and which item classes are associated with higher or lower profits compared to the average?


from sklearn.preprocessing import OneHotEncoder
import statsmodels.api as sm

# --- Step 1: Prepare Data ---
df = df_inf.copy()
target = df['log_profit']
col = 'item_class_code'

# Ensure column is string
df[col] = df[col].astype(str)

# Keep only top 10 most frequent values; group the rest as "Other"
top10 = df[col].value_counts().nlargest(10).index.tolist()
df['item_class_group'] = df[col].apply(lambda x: x if x in top10 else 'Other')

# --- Step 2: One-Hot Encode ---
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(df[['item_class_group']])
encoded_cols = encoder.get_feature_names_out(['item_class_group'])
X_df = pd.DataFrame(X_encoded, columns=encoded_cols, index=df.index)

# --- Step 3: Add Constant ---
X_final = sm.add_constant(X_df)

# --- Step 4: Fit OLS Model ---
model = sm.OLS(target, X_final, missing='drop').fit()
print(f"\nFitting OLS model for '{col}' (Top 10 + Other), using {X_final.shape[1] - 1} predictors (excluding const).\n")
print(model.summary())
```
